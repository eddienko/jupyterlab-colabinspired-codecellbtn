```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normpdf(x, mean=0, std=1):
    y = (x - mean) / std
    f = np.exp(-y * y / 2) / np.sqrt(2 * np.pi) / std
    return f
```

```python
x = np.arange(-5, 5, 0.01)
plt.plot(x, normpdf(x, 0, 1), label='$\sigma$ = 1')
plt.plot(x, normpdf(x, 0, 2), label='$\sigma$ = 2')
plt.plot(x, normpdf(x, 0, 4), label='$\sigma$ = 3')
plt.legend()
plt.show()
```
